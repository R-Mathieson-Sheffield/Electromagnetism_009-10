## Electrical Power

Power is defined as *the rate at which work is done*. As work is measured in Joules it is a unit of energy, as such the previous statement can be defined mathematically as:

$$ P = \frac{\Delta E}{\Delta t}, $$

where $P$ is the power, $E$ is energy and $t$ is time. We may also recall that voltage or *electric potential*, $V$, is defined as the electrical energy per unit charge:

$$ V = \frac{\Delta E}{\Delta Q}. $$

If we are to divide both sides of the equation above by $\Delta t$, we get:

$$ \frac{V}{\Delta t} = \frac{\Delta E}{\Delta Q \Delta t}, $$

multiplying the $\Delta t$ term from the left hand side to the right hand side gives:

$$ V = \frac{\Delta E \Delta t}{\Delta Q \Delta t} = \frac{\Delta E}{\Delta t}\frac{\Delta t}{\Delta Q}.$$

Above we established that the $\frac{\Delta E}{\Delta t}$ is simply equal to the power, and $\frac{\Delta t}{\Delta Q}$ is simply the reciprocal of the current. As such the expression above may be expressed as:

$$ V = \frac{P}{I}, $$

which gives us our expression for power:

$$ P = IV. $$ 

Power has the units of J$\,$s$^{-1}$, which is known as the **Watt, W**.

Alternative forms of this equation can be derived by introducing Ohm's law, $V = IR$:

$$ P = IV,$$

$$ \therefore P = I(IR), $$

$$ \therefore P = I^2 R, $$

and 

$$ P = IV, $$

$$ \therefore P = \left(\frac{V}{R}\right) V, $$

$$ \therefore P = \frac{V^2}{R}. $$


```{admonition} Important

Electrical power is measured in **Watts**, **W**, and can be calculated using the following equations:

$$ P = IV = I^2 R = \frac{V^2}{R}. $$

```

---
## Efficiency

![Bulb](Figures/Bulb.jpeg)

When we supply electrical power to a device, we expect it to use that energy to do something useful. For example, when I turn on a light I expect that electrical energy to be converted into light energy to illuminate a space. This light energy is useful energy.

Some of that electrical energy will also be converted into other types of energy. Taking my light as an example, should I be using an old incandescent bulb, the vast majority of the electrical energy will be converted into heat energy via resistive heating in the tungsten filament. If my aim is to illuiminate my room this is not useful energy.

The efficiency, $\eta$, of a device is the measure of how much energy it converts into useful energy:

$$ \eta = \frac{\text{useful energy out}}{\text{total energy in}}. $$

If we have an incandescent bulb supplied with $60\,$ J of electrical energy and it produces $10\,$ J of light energy, the efficiency of the bulb is:

$$ \eta = \frac{10 \, J}{60\, J} = 0.167, $$

therefore the bulb has an efficiency of 16.7%. 

This calculation can also be flipped on its head. Let's assume that this bulb is actually being used to heat up a chicken coop meaning now the useful energy is the thermal energy it kicks out, which assuming there are no other energy losses would be equal to $50\, $J:

$$ \eta = \frac{50 \, J}{60\, J} = 0.833, $$

therefore the bulb has an efficiency of 83.3% when used as a heater.

```{admonition} Important

Efficiency is the ratio of useful energy (or power) output to the total amount of energy (or power) supplied to a device:

$$ \eta = \frac{\text{useful energy out}}{\text{total energy in}}. $$

As efficiency is a ratio it has no units, but can be expressed as a percentage by mutiplying by 100.
```

```{admonition} Q1a Solution 
:class: dropdown

To determine the total power of the circuit we must first determine the total equivalent resistance of the circuit. The circuit contains both resistors arranged in series and in parallel. We first must determine the equivlent resitance of the resistors arranged in series:

$R_S = R_2 + R_3 + R_4, $

$R_S = 4.00\,\Omega + 6.00\,\Omega + 8.00\,\Omega, $

$R_S = 18.0\,\Omega. $

We can then determine the total equivalent resistance of the circuit using our resistors in parallel rule:

$\frac{1}{R_T} = \frac{1}{R_1} + \frac{1}{R_S}, $

$\frac{1}{R_T} = \frac{1}{12.0\, \Omega} + \frac{1}{18.0\, \Omega} = \frac{5}{36} \, \Omega^{-1}, $

$ R_T = \frac{36}{5} \Omega. $

Now that we know the total equivalent resistance of the circuit, we can determine the power it outputs:

$ P = \frac{V^2}{R} $

$ P = \frac{(15.0\,V)^2}{\frac{36}{5}\,\Omega} $

$ P = 31.3 W $

```

```{admonition} Q1b Solution 
:class: dropdown

We know that the voltage drop across resistor R1 is simply equal to the voltage of the supply as they are arranged in parallel. Therefore we can simply determine the power as:

$ P = \frac{V^2}{R} $

$ P = \frac{(15.0\, V)^2}{12.0\, \Omega} $

$ P = 18.8 \, V$
```


```{admonition} Q1c Solution 
:class: dropdown

To determine the power output by resistor R4 we either need to know the voltage drop across it or the current flowing through it. Let's determine what the current flowing through it is. We can calculate the total current flowing through the circuit using Ohm's law as we know the total equivalent resistance of the circuit:

$ I_T = \frac{V}{R_T}, $

$ I_T = \frac{15.0\, V}{\left(\frac{36}{5} \Omega.\right)}, $

$ I_T = \frac{25}{12}\, A. $

Logically, the current is split at the junction between the R1 and the rest of the resistors. We need to therefore determine how much current is flowing through each section of the circuit. The current flowing through R1 is calculated again using Ohm's law:

$ I_{R1} = \frac{V}{R_1}, $ 

$ I_{R1} = \frac{15.0 \, V }{12.0\, \Omega}, $

$ I_{R1} = \frac{5}{4}\, A $

Therefore the remaining current flowing through the rest of the resistors is calculated as:

$ I = I_T - I_{R1,} $

$ I = \frac{25}{12}\, A - \frac{5}{4} \, A, $

$ I = \frac{5}{6} \, A $

We can now determine the power dissipated by R4:

$ P = I^2 R, $

$ P = \left(\frac{5}{6} \, A\right)^2 \cdot 8.00 \,\Omega, $

$ P = 5.56 \, W. $

```

---
### Question 2

It takes $15000\,$kJ of energy to increase the temperature of a room by 10ËšC.

a) If a 2000W heater with a 100% efficiency is used to increase the temperature of the room, how long does it take for the temperature of the room to increase?

b) The heater is connected to a $230\,$V power supply. What is the current being supplied to the heating element in the heater?

c) The heating element itself is made of a metal wire. What is the resistance of this wire?

d) The heater is replaced with a another that has an efficiencty of $\eta = 0.75$. How long does it take for this heater to heat the room by the same amount?

```{admonition} Q2a Solution 
:class: dropdown

For this question we just need to remember that power is the rate at which work is done, and that power is measured in watts (joules per second):

$ P = \frac{\Delta E}{\Delta t}, $

$ \Delta t = \frac{\Delta E}{P}, $

$ \Delta t = \frac{15000 \times 10^3\, J}{2000\, W}, $

$ \Delta t = 7500\, s. $
```


```{admonition} Q2b Solution 
:class: dropdown

If we know the power and the voltage supplied, we can determine what the current is easily:

$P = IV $

$ I = \frac{P}{V}, $

$ I = \frac{2000\, W}{230\, V}, $

$ I = \frac{200}{23} \, A = 8.7\, A. $
```


```{admonition} Q2c Solution 
:class: dropdown

Again if we know the power and the voltage, we can easily determine the resistance:

$ P = \frac{V^2}{R}, $

$ R = \frac{V^2}{P}, $

$ R = \frac{\left(230\, V\right)^2}{2000\, W}. $

$ R = 26.5 \, \Omega, $

$ R = 27 \, \Omega \text{(2 sig figs)}. $

Alternativlely, as we knew the voltage and we determined the current in the previous question, we could have simply used Ohm's law:

$ R = \frac{V}{I}, $

$ R = \frac{230\, V}{\left(\frac{200}{23}\right) \, A}, $

$ R = 26.5 \, \Omega, $

$ R = 27 \, \Omega \text{(2 sig figs)}. $
```


```{admonition} Q2d Solution 
:class: dropdown

We know that if the efficiency of the heater is lower, less electrical energy is being converted into heat energy and it is going to take longer for the heater to heat the room. We can calculate how long by dividing the time it took for the 100% efficient heater to increase the temperature of the room by the efficiency of the second heater:

$ t = \frac{ t_{100} }{\eta}, $

$ t = \frac{7500 \,s }{0.75}, $

$ t = 10000\, s .$
```

---
### Question 3

A handheld fan is supplied with $12\,V$ of EMF and a $0.5\, A$ current. In 30 seconds only $120\,J$ of kinetic energy was generated to turn the fan blades. What is the efficiency of the fan?

```{admonition} Q3 Solution 
:class: dropdown

To determine the efficiency, we need to determine the total amount of energy supplied to the fan in a 30 second period. We can do this by determining the power:

$ P = IV $ 

$ P = 0.5\, A \times 12\, V$

$ P = 6 \,W = 6\, J\,s^{-1}. $

If the power input into the fan in one second is $6\, J$, then the total amount of energy supplied in a 30 second period is:

$ P = \frac{E}{t} ,$

$ E = P t ,$

$ E = 6\, W \times 30\, s,$

$ \Delta E = 180 \, J. $

To determine the efficiency we simply divide the amount of useful energy by the total energy input:

$ \eta = \frac{120\, J}{180\, J}, $

$ \eta = 0.67. $

The efficiency of the fan is therefore 67%.
```
